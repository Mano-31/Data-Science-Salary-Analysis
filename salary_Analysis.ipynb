```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("ds_salaries.csv")

df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df = df.drop("Unnamed: 0", axis=1)

df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

# Analysis 1: Most Common Job Roles

```python
df["job_title"].value_counts().head(10)
```

```python
import matplotlib.pyplot as plt

df["job_title"].value_counts().head(10).plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top 10 Job Roles")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.show()
```

# Analysis 2: Salary by Experience Level

```python
df.groupby("experience_level")["salary_in_usd"].mean()
```

```python
df.groupby("experience_level")["salary_in_usd"].mean().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Average Salary by Experience Level")
plt.ylabel("Salary (USD)")
plt.show()
```

# Findings

- The dataset contains no missing values.
- Several data-related job roles are represented.
- Salary varies significantly across experience levels.
- Senior-level positions tend to earn higher salaries.

```python
df["job_title"].value_counts().head(10)
```

# Analysis 3: Highest Paying Job Roles

```python
top_salary_jobs = df.groupby("job_title")["salary_in_usd"].mean().sort_values(ascending=False)

top_salary_jobs.head(10)
```

```python
top_salary_jobs.head(10).plot(
    kind="bar",
    figsize=(12,6)
)

plt.title("Top 10 Highest Paying Data Jobs")
plt.xlabel("Job Title")
plt.ylabel("Average Salary (USD)")
plt.show()
```

# Analysis 4: Salary by Company Size

```python
df.groupby("company_size")["salary_in_usd"].mean()
```

```python
df.groupby("company_size")["salary_in_usd"].mean().plot(
    kind="bar",
    figsize=(8,5)
)

plt.title("Average Salary by Company Size")
plt.ylabel("Salary (USD)")
plt.show()
```

# Analysis 5: Remote Work Trends

```python
df["remote_ratio"].value_counts()
```

```python
df["remote_ratio"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    figsize=(6,6)
)

plt.title("Remote Work Distribution")
plt.ylabel("")
plt.show()
```

```python
df["company_location"].value_counts().head(10)
```

```python
df["company_location"].value_counts().head(10).plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Top Hiring Countries")
plt.ylabel("Number of Jobs")
plt.show()
```
